```python
import turicreate as tc
```

# Bib Detector

```python
# Load the data
data = tc.SFrame('rbnr_tiny.sframe')
```

```python
# Make a train-test split
train_data, test_data = data.random_split(0.8)
```

```python
# Create a model
bd_model = tc.object_detector.create(train_data)
```

```python
# Save predictions to an SArray
predictions = bd_model.predict(test_data)
```

```python
# Evaluate the model and save the results into a dictionary
metrics = bd_model.evaluate(test_data)
metrics
```

```python
# Save the model for later use in Turi Create
bd_model.save('../config/rbn_detector.model')
```

```python
# Export for use in Core ML
bd_model.export_coreml('../config/rbn_detector.mlmodel')
```
